```python
#This cell will combine several .csvs from main.py into 1 .csv
import pandas as pd
import glob

# Specify the folder containing the CSV files
folder_path = #enter your filepath here

# List all CSV files in the folder
csv_files = glob.glob(folder_path + '/*.csv')

# Initialize an empty list to store the DataFrames
dataframes = []

# Loop through the CSV files
for file in csv_files:
    # Read the current CSV file into a DataFrame
    df = pd.read_csv(file)
    
    # Ensure the columns are in the correct order
    df = df[['company', 'job_title', 'link', 'salary', 'posted_date']]
    
    # Append the DataFrame to the list
    dataframes.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('all_listings.csv', index=False)
```

```python
#this cell groups job listings by companies, 
import pandas as pd

# Read the input CSV file
input_file = "leads.csv"
df = pd.read_csv(input_file)

# Assuming you have a DataFrame named df, and the first column's name is 'company', and the third column's name is 'third_column', and the fifth column's name is 'date'

# Convert the 'date' column to datetime objects
df['date'] = pd.to_datetime(df['posted_date'], format='%m-%d-%Y')

# Group the DataFrame by the 'company' column, then use the 'agg' function to aggregate the data
result = df.groupby('company').agg(
    count=('company', 'count'),
    link=('link', 'first'),  # Replace 'first' with the desired aggregation function, e.g., 'mean', 'sum', 'max', etc.
    most_recent_listing=('date', 'max')
)

# Reset the index
result = result.reset_index()
result.to_csv('hotlist.csv', index=False)
```
